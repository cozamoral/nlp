```python
import pandas as pd
import re
import json
import nltk
nltk.download('punkt')
from gensim.summarization import summarize
from nltk import tokenize
from rouge import Rouge
```

```python
dataset = pd.read_json('/content/drive/My Drive/Lab/dataset_43428_1.txt')
dataset
```

```python
def text_normalization(row):
    text = row[0]
    text = text.lower()
    text = re.sub(r'\n|\r', ' ', text)
    text = re.sub(r' +', ' ', text)
    text = text.strip()
    return text
```

```python
dataset['normalized']= dataset.apply(lambda row: text_normalization(row), axis=1)
```

```python
def text_summary(row):
    text = row['normalized']
    count = len(tokenize.word_tokenize(text))
    sent = tokenize.sent_tokenize(text)
    if len(sent) == 1:
        return text
    if count <= 2000:
        x = summarize(text, ratio=0.2, split=False)
    elif count >= 2000:
        x = summarize(text, word_count=300, split=False)
    return x
```

```python
dataset['summary'] = dataset.apply(lambda row: text_summary(row), axis=1)
dataset['summary']
```

```python
dataset
```

```python
def word_count_text(row):
    text = row[0]
    word_list = tokenize.word_tokenize(text)
    return len(word_list)

def word_count_summary(row):
    text = row['summary']
    word_list = tokenize.word_tokenize(text)
    return len(word_list)
```

```python
dataset['word_count_text'] = dataset.apply(lambda row: word_count_text(row), axis=1)
```

```python
dataset['word_count_summary'] = dataset.apply(lambda row: word_count_summary(row), axis=1)
```
